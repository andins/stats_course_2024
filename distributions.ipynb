```python
eval_range = np.linspace(start=-4, stop=4, num=100)
pdf = sps.norm.pdf(x=eval_range, loc=1, scale=1)
plt.plot(eval_range, pdf)
plt.xlabel("x")
plt.ylabel("PDF")
plt.title("A boring Gaussian distribution")
```

```python
eval_range = np.linspace(0, 5, 100)
pdf = sps.expon.pdf(x=eval_range, scale=.5)  # scale is the inverse of the rate parameter (lambda)
plt.plot(eval_range, pdf)
plt.xlabel("x")
plt.ylabel("PDF")
plt.title("Exponential Distribution")
```

```python
shape = 2  # shape parameter
scale = 10  # scale parameter
eval_range = np.linspace(0, 15, 100)
pdf = sps.gamma.pdf(x=eval_range, a=shape, scale=scale)
plt.plot(eval_range, pdf, color='orange')
plt.xlabel("x")
plt.ylabel("PDF")
plt.title("Gamma Distribution (shape=2, scale=1)")
```

```python
a, b = 2, 5  # shape parameters
eval_range = np.linspace(0, 1, 100)
pdf = sps.beta.pdf(x=eval_range, a=a, b=b)
plt.plot(eval_range, pdf, color='green')
plt.xlabel("x")
plt.ylabel("PDF")
plt.title("Beta Distribution (a=2, b=5)")
```

```python
p = 0.3  # probability of success
x = [0, 1]
pmf = sps.bernoulli.pmf(x, p)
plt.bar(x, pmf, color='purple', width=0.1)
plt.xticks(x)
plt.xlabel("x")
plt.ylabel("PMF")
plt.title("Bernoulli Distribution (p=0.3)")
```

```python
n = 10  # number of trials
p = 0.5  # probability of success
x = np.arange(0, n + 1)
pmf = sps.binom.pmf(x, n, p)
plt.bar(x, pmf, color='red', width=0.1)
plt.xticks(x)
plt.xlabel("x")
plt.ylabel("PMF")
plt.title("Binomial Distribution (n=10, p=0.5)")
```

```python
mu = 3  # average rate (lambda)
x = np.arange(0, 15)
pmf = sps.poisson.pmf(x, mu)
plt.bar(x, pmf, color='cyan', width=0.5)
plt.xticks(x)
plt.xlabel("x")
plt.ylabel("PMF")
plt.title("Poisson Distribution (λ=3)")
```

Transformaciones

```python
x = np.random.normal(loc=10, scale=3, size=1000)
plt.hist(x, bins=30);
```

```python
x_centered = x - x.mean()
plt.hist(x_centered, bins=30);
```

```python
center = x.mean()
scale = x.std()
x_centered_scaled = (x - center) / scale
plt.hist(x_centered_scaled, bins=30);
```

```python
def center_scale_transform(x, center=0, scale=1):
    return (x - center) / scale
```
